```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/barchart_dataset.csv')
print(data.head())
print(data.tail())
```

```python
ages = data['AGE']


plt.figure(figsize=(10, 6))
plt.hist(ages, bins=10, edgecolor='black')
plt.title('Distribution of Ages in the Population')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


```

```python
genders = data['GENDER']
gender_counts = genders.value_counts()

plt.figure(figsize=(10, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Distribution of Genders in the Population')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
